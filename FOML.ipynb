```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


url="https://raw.githubusercontent.com/Balamurugan2007/FOML/refs/heads/main/IRIS.csv"
df = pd.read_csv(url)

print(df.head())

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for feature in features:
    plt.figure(figsize=(6, 3))
    sns.scatterplot(x=df[feature], y=[0]*len(df), hue=df['species'])
    plt.title(f'Univariate Analysis - {feature}')
    plt.xlabel(feature)
    plt.ylabel('')
    plt.yticks([])
    plt.show()

pairs = [('sepal_width', 'petal_width'),
         ('sepal_length', 'petal_length')]
for x_feature, y_feature in pairs:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=df[x_feature], y=df[y_feature], hue=df['species'])
    plt.title(f'Bivariate Analysis - {x_feature} vs {y_feature}')
    plt.show()


df_multi = df # Removed df.drop(columns=['Id'])
sns.pairplot(df_multi, hue='species', diag_kind='kde')
plt.suptitle("Multivariate Analysis", y=1.02)
plt.show()
```
