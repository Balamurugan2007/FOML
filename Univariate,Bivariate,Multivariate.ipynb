```python
import pandas as pd
file_path = '/content/drive/MyDrive/FOML/IRIS.csv'
df = pd.read_csv(file_path)
print(df.head())
```

```python
data = {
'Person': list (range (1, 16)),
'Height cm': [160, 165, 158, 170, 172, 168, 175, 162, 160, 168, 173, 165, 166, 164, 176],
'Weight kg': [45, 50, 52, 60, 65, 68, 72, 74, 77, 80, 85, 58, 62, 49, 90],
'Category': ['Slim', 'Slim','Slim','Fit','Fit','Fit','Fit','Obese','Obese','Obese','Obese', 'Fit', 'Fit', 'Slim','Obese' ] }
```

```python
df=pd.DataFrame(data)
```

```python
plt.figure (figsize=(8, 5))
sns.scatterplot (x=df['Weight kg'], y=[0]*len (df),hue=df['Category'], s=100)
plt.yticks ([])
plt.xlabel ("Weight (kg) ")
plt.title("Univariate Analysis of Weighttegory")
plt.legend (title="Category")
plt.grid(True)
plt.show()


```

```python
plt.figure (figsize=(10, 6))
sns.scatterplot (data=df, x='Height cm', y='Weight kg',hue='Category', s=100)
plt.xlabel("Height (cm) ")
plt.ylabel("Weight (kg) ")
plt.title("Bivariate Analysis of Height and Weight by Category")
plt.legend (title="Category")
plt.grid(True)
plt.show()

```

```python
df['Age'] = [25, 30, 22, 35, 40, 32, 45, 50, 55, 60, 65, 28, 33, 26, 70]

sns.pairplot(df, hue='Category',
             vars=['Height cm','Weight kg','Age'])

plt.suptitle("Multivariate Analysis of Height, Weight, and Age by Category", y=1.02)
plt.show()
```
